```python
def one_expr(n, o_func, o_thr, adj_np=None, u_np=None):
    if adj_np is None:    
        adj_res = get_random_adj(n, o_func, o_thr, p=0.4)
        adj_np = adj_res['adj']
        u_np = adj_res['u']
        # v_np = adj_res['v']
        # o_np = adj_res['o']
    
    # print("adj:", repr(adj_np))
    # print("u:", repr(u_np))
    # print("v:", v_np)
    # print("o:", o_np)

    tic = time.time()
    a_cause_hp = next(hp_cause(adj_np, u_np, o_func, o_thr))
    toc = time.time()
    delta_hp = toc - tic

    tic = time.time()
    a_cause_hp_nn = hp_cause_nn(adj_np, u_np, o_func, o_thr)
    toc = time.time()
    time_hp_nn = toc - tic

    return dict(
        adj=adj_np,
        u=u_np,
        #v=v_np,
        #o=o_np,
        a_cause_hp=a_cause_hp,
        delta_hp=delta_hp,
        a_cause_hp_nn=a_cause_hp_nn,
        delta_hp_nn=time_hp_nn,
    )
```

```python
res = one_expr(n, o_func, o_thr)
```

```python
results = defaultdict(list)
for n in [10]:
    for _ in tqdm(range(4)):  
        res = one_expr(n, o_func, o_thr)
        results[n].append(res)

import pickle
import os
import datetime
os.makedirs('results', exist_ok=True)
current_time = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
with open(f'results/results_{current_time}.pkl', 'wb') as f:
    pickle.dump(results, f)
#print(results)
print(f"Results saved to results/results_{current_time}.pkl")
```
